```python
import pandas as pd
import random
import seaborn as sns
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

#algoritmos
from sklearn.linear_model import LinearRegression #Regresion Lineal
from sklearn.linear_model import  LogisticRegression #Regresion logistica
from sklearn.tree import DecisionTreeClassifier#Arboles de decision
from sklearn.tree import plot_tree
from sklearn.ensemble import RandomForestClassifier #Bosques aleatorios
from sklearn.metrics import mean_squared_error # calcula el error cuadratico medio
```

```python
import random

# Generación del dataset
dataset = {
    "EmpleadoID": [i for i in range(1, 101)],
    "Edad": [random.randint(22, 60) for _ in range(100)],
    "Anios_Experiencia": [random.randint(0, 40) for _ in range(100)],
    "Nivel_Educacion": [random.choice(["Secundario", "Universitario", "Posgrado"]) for _ in range(100)],
    "Puntaje_Desempeno": [round(random.uniform(50, 100), 2) for _ in range(100)],
    "Certificaciones": [random.randint(0, 5) for _ in range(100)],
    "Horas_Extra": [random.randint(0, 20) for _ in range(100)],
    "Departamento": [random.choice(["Ventas", "IT", "RRHH", "Finanzas", "Producción"]) for _ in range(100)],
    "Genero": [random.choice(["Masculino", "Femenino"]) for _ in range(100)],
    "Estado_Civil": [random.choice(["Soltero", "Casado", "Divorciado"]) for _ in range(100)],
    "Salario": [random.randint(30000, 120000) for _ in range(100)],
    "Satisfaccion_Laboral": [random.randint(1, 5) for _ in range(100)],  # Escala de 1 a 5
    "Evaluacion_Anual": [random.randint(1, 10) for _ in range(100)],  # Escala de 1 a 10
    "Promedio_Horas_Trabajo": [random.randint(30, 60) for _ in range(100)],
    "Promovido": [random.choice([0, 1]) for _ in range(100)]  # 0: No promovido, 1: Promovido
}

# Ejemplo de cómo convertirlo en un DataFrame si usas pandas
import pandas as pd
df = pd.DataFrame(dataset)
```

```python
label_encoder = LabelEncoder()

# Transformar cada columna
df["Nivel_Educacion"] = label_encoder.fit_transform(df["Nivel_Educacion"])
df["Departamento"] = label_encoder.fit_transform(df["Departamento"])
df["Genero"] = label_encoder.fit_transform(df["Genero"])
df["Estado_Civil"] = label_encoder.fit_transform(df["Estado_Civil"])
```

```python
df
```

```python
escala =  MinMaxScaler(feature_range=(0,1)) #Define el rango al que se escalarán los datos. En este caso, los valores se escalan entre 0 y 1.
normado = escala.fit_transform(df) #Calcula el valor mínimo y máximo de cada columna en el DataFrame original (df)
df_normado = pd.DataFrame(data=normado , columns= df.columns)# verifica que los nombres de las columnas sean iguales a las df orginal
```

```python
df_normado
```

```python
X = df_normado.drop('Promovido' , axis=1)
```

```python
y = df_normado['Promovido']
```

```python
X_entrena,X_prueba , y_entrena , y_prueba = train_test_split(X,y , train_size=0.8, random_state=42)
```

```python
modelo = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
modelo.fit(X_entrena,y_entrena)
```

```python
modelo.score(X_prueba,y_prueba)
```
